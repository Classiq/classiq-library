```python
from math import pi

import matplotlib.pyplot as plt
from qiskit import *
```

```python
x = [
    0.9209112552070406,
    0.8204347455451488,
    0.7375553620097723,
    0.9881416472095654,
    0.8157508399791834,
    0.6085057148243024,
]
y = [
    0.21332099592517406,
    0.47005158517477286,
    0.5195365169616781,
    0.2946539313446742,
    0.254401699122084,
    0.9010739339091007,
]
# print(x)
# print(y)
```

```python
# n is the number of points in the dataset


def distance_estimator(n):
    # theta_list for the x values
    theta_list = [((x + 1) * pi / 2) for x in x]
    # phi_list for the y values
    phi_list = [((y + 1) * pi / 2) for y in y]
    qreg = QuantumRegister(3, "qreg")
    creg = ClassicalRegister(1, "creg")
    qc = QuantumCircuit(qreg, creg, name="qc")
    backend = Aer.get_backend("qasm_simulator")
    qreg = QuantumRegister(3, "qreg")
    creg = ClassicalRegister(1, "creg")
    qc = QuantumCircuit(qreg, creg, name="qc")
    backend = Aer.get_backend("qasm_simulator")
    results_list = []
    for i in range(n):
        for j in range(1, n - i):
            qc.h(qreg[2])
            qc.u3(theta_list[i], phi_list[i], 0, qreg[0])
            qc.u3(theta_list[i + j], phi_list[i + j], 0, qreg[1])
            qc.cswap(qreg[2], qreg[0], qreg[1])
            qc.h(qreg[2])
            qc.measure(qreg[2], creg[0])
            qc.reset(qreg)
            job = execute(qc, backend=backend, shots=1024)
            result = job.result().get_counts(qc)
            results_list.append(result["1"])

    # seperating list into sublists
    first = 0
    end = n - 1
    done = False
    arr = results_list
    i = 1
    sub_arrays = []
    while not done:
        Li = arr[first:end]
        first = end
        i += 1
        end = min(end + n - i, len(arr))
        done = len(Li) == 1
        sub_arrays.append(Li)

    # creating a list of alphabetic chars represnting the points
    Alphabetlist = [chr(i) for i in range(65, 65 + n)]

    # creating a dict based on
    dist_dict = dict(zip(Alphabetlist, sub_arrays))

    # finding the lowest value and refers it to the point it was calculated from

    minVal = 300
    lowInitial = ""
    for initials, dists in dist_dict.items():
        lowestScore = min(dists)
        if lowestScore < minVal:
            minVal = lowestScore
            lowInitial = initials

    return dist_dict, lowInitial, minVal
```
