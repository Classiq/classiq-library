### **Implementation:**

> #### **Operator W**

$$ W=\sum_{j\in V}^{N} \left| j\right\rangle \left\langle j\right|  \otimes \left( 2 \left| \partial_j\right\rangle \left\langle \partial_j\right| - I\right)$$

1. **Initialization of the probability vector**:
    * For each vertex $j$ in the set of vertices $V$, initialize a probability vector `prob` with zeros. This vector will later be used to define the state $\left| \partial_j\right\rangle$.<br>
3. **Setting transition probabilities**:
    * For the first and last vertices, the transition probabilities are set to move to their only adjacent vertices with probability 1.
    * For all other vertices $j$, the probabilities are set to 0.5 for moving to the next vertex $(j+1) \% 16$ and the previous vertex $(j-1) \% 16$.<br>
4. **Quantum circuit for state preparation**:
    * Apply the `inplace_prepare_state` function to create the superposition state $\left| \partial_j\right\rangle$ based on the probabilities defined in the `prob` vector.<br>
5. **Diffusion operation**:
    * Implement the operation $2 \left| \partial_j\right\rangle \left\langle \partial_j\right| - I$ using the `zero_diffuzer` function, which utilizes an auxiliary qubit in the state $\left| - \right\rangle$ and an oracle to perform the phase flip, achieving the diffusion operation.<br>
6. **Controlled application**:
    * Use a controlled operation to apply the above quantum operations only to the respective vertex $j$, ensuring that each vertex state is updated independently based on its adjacency.



> #### **Operator S**

$$S=\sum_{(j,k)\in E}^{n} \left| j,k\right\rangle \left\langle j,k\right| $$

1. **Edge oracle**:
    * Define an oracle `edge_oracle` to check adjacency between any two vertices $j$ and $k$. This oracle sets a result qubit if $j$ and $k$ are adjacent.<br>
2. **Bitwise SWAP**:
    * Implement a bitwise swap operation `bitwise_swap` which swaps the quantum states of two vertices $j$ and $k$ if they are found to be adjacent by the edge oracle.<br>
3. **Controlled Swap**:
    * Use a control operation to conditionally execute the bitwise swap only if the result from the `edge_oracle` indicates that the vertices $j$ and $k$ are adjacent.<br>

```python
# Sets the quantum register size to 4 qubits, corresponding to 2^4 = 16 possible states.
size = 4

@qfunc
def prepare_minus(x: QBit):
    X(x)  # Applies the Pauli-X gate to the qubit x
    H(x)  # Applies the Hadamard gate to qubit x, creating superposition with phase -1
    
@qfunc
def diffuzer_oracle(aux: Output[QNum],x:QNum):
    # Applies a conditional phase flip when x is not zero
    aux^=(x!=0)

@qfunc
def zero_diffuzer(x: QNum):
    aux = QNum('aux')
    allocate(1,aux)  # Allocate 1 auxiliary qubit
    # Uses the within_apply structure to apply the prepare_minus and diffuzer_oracle functions in a specific order
    within_apply(compute=lambda: prepare_minus(aux), action=lambda: diffuzer_oracle)

def W_iteration(i:int,vertices: QNum, adjacent_vertices:QNum):
    prob = [0] * 16  # Initializes a probability vector for 16 vertices with all zeros
    # Transition probability for the first vertex is set to 1
    if i == 0:
        prob[1] = 1 
    # Transition probability for the last vertex is set to 1
    elif i == 15:
        prob[14] = 1
    else:
        prob[(i + 1) % 16] = 0.5  # Probability of moving to the next vertex
        prob[(i - 1) % 16] = 0.5  # Probability of moving to the previous vertex
        
    print(f'State={i}, prob vec ={prob}')  # Outputs the current state and its probability vector
    
    # Applies the transformation only when the control condition (vertices == i) is true
    control(ctrl=vertices==i,
            operand=lambda: within_apply(
              compute= lambda: inplace_prepare_state(probabilities=prob, bound=0.01, target=adjacent_vertices),
              action= lambda: zero_diffuzer(adjacent_vertices)))
    
@qfunc 
def W_operator(vertices:QNum, adjacent_vertices: QNum):
    # Iterates over all possible states for the vertices
    for i in range(2**size):
      W_iteration(i,vertices,adjacent_vertices)   # Applies the W_iteration function to each vertex
```

```python
@qfunc
def edge_oracle(res: Output[QBit], vertices: QNum, adjacent_vertices: QNum):
    # This oracle function sets the result qubit 'res' based on whether the vertices are adjacent in the line
    # It checks if the difference between the indices of 'vertices' and 'adjacent_vertices' is 1
    res |= ((vertices - adjacent_vertices) == (1 or -1))
    
@qfunc 
def bitwise_swap(x: QArray[QBit], y:QArray[QBit]):
    # Executes a SWAP gate between each corresponding pair of qubits in arrays 'x' and 'y'
    # The 'repeat' function iterates over the length of the array 'x' to perform SWAPs
    repeat(count= x.len, iteration= lambda i: SWAP(x[i],y[i]))

@qfunc 
def S_operator(vertices:QNum, adjacent_vertices: QNum):
    # Defines a shift operator that conditionally swaps 'vertices' and 'adjacent_vertices' if they are adjacent
    res = QNum('res')  # Allocates a qubit to store the result of the adjacency check
    edge_oracle(res,vertices,adjacent_vertices)  # Calls the edge oracle to determine adjacency
    # Conditionally executes a bitwise swap if the result qubit 'res' indicates adjacency (res == 1)
    control(ctrl= res==1, operand= lambda: bitwise_swap(vertices,adjacent_vertices))

@qfunc 
def main(vertices: Output[QNum], adjacent_vertices: Output[QNum]):
    # The main function orchestrates the initialization, operation, and interaction of the quantum walk
    allocate(size, vertices)  # Allocates a quantum register for 'vertices' based on the predefined 'size'
    hadamard_transform(vertices)  # Applies a Hadamard transform to all qubits in 'vertices' to create superposition
    allocate(size, adjacent_vertices)  # Similarly allocates a quantum register for 'adjacent_vertices'

    W_operator(vertices, adjacent_vertices)  # Calls the walk operator on the vertices
    S_operator(vertices, adjacent_vertices)  # Calls the shift operator on the vertices
```

```python
# Create the quantum model
quantum_model = create_model(main)
```

```python
# Synthesize the quantum program
quantum_program = synthesize(quantum_model)
```
