```python
num_qubits = 2
```

```python
@qfunc
def circuit(qarr: QArray) -> None:
    # pass
    H(qarr[0])
    CX(qarr[0], qarr[1])

@qfunc(generative=True)
def main(qarr: Output[QArray]) -> None:
    allocate(num_qubits, qarr)
    circuit(qarr)
```

```python
qmod = create_model(main)
qprog = synthesize(qmod)
show(qprog)
```

```python
snapshots_range = [10, 50, 100, 250]
distances = [1.68, 0.69, 0.46, 0.37]
plt.plot(snapshots_range, distances)
plt.title("Distance between Ideal and Shadow Bell States")
plt.xlabel("Number of Snapshots")
plt.ylabel("Distance")
plt.show()
```
