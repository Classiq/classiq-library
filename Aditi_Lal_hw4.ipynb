# **Quantum Walk Operator for a Line with 16 Nodes**
**Introduction** \
In this notebook, we will design a quantum walk operator for a line graph consisting of 16 nodes. Quantum walks are the quantum analogs of classical random walks and can be used to model various quantum phenomena.

```python
import numpy as np
import matplotlib.pyplot as plt
```

# **Define the Quantum Walk Operator**
For a line with 16 nodes, we will create a unitary operator that represents the quantum walk. The operator will be defined based on the adjacency matrix of the line graph.

```python
def create_line_operator(n):
    """Create the quantum walk operator for a line with n nodes."""
    # Initialize the operator as a zero matrix
    operator = np.zeros((n, n), dtype=complex)

    # Fill the operator based on the adjacency of the line graph
    for i in range(n):
        if i > 0:
            operator[i, i-1] = 1  # Connect to the left neighbor
        if i < n - 1:
            operator[i, i+1] = 1  # Connect to the right neighbor

    # Normalize the operator
    operator = operator / np.sqrt(2)
    return operator
```

# **Visualize the Operator**
We can visualize the quantum walk operator using a heatmap.

```python
def plot_operator(operator):
    plt.figure(figsize=(8, 6))
    plt.imshow(np.abs(operator), cmap='hot', interpolation='nearest')
    plt.colorbar()
    plt.title("Quantum Walk Operator for Line Graph")
    plt.xlabel("Node")
    plt.ylabel("Node")
    plt.show()

# Create and plot the operator for 16 nodes
n = 16
line_operator = create_line_operator(n)
plot_operator(line_operator)
```

# **Simulate the Quantum Walk**
We can now simulate the quantum walk using the created operator.


```python
def simulate_walk(operator, initial_state, steps):
    """Simulate the quantum walk."""
    state = initial_state
    states = [state]

    for _ in range(steps):
        state = operator @ state  # Apply the operator
        states.append(state)

    return states

# Initial state: starting at the first node
initial_state = np.zeros(n)
initial_state[0] = 1  # Start at node 0

# Simulate the walk for a number of steps
steps = 10
walk_states = simulate_walk(line_operator, initial_state, steps)

# Plot the state distribution over time
for i, state in enumerate(walk_states):
    plt.plot(state.real, label=f'Step {i}')

plt.title("Quantum Walk Simulation on a Line Graph")
plt.xlabel("Node")
plt.ylabel("Probability Amplitude")
plt.legend()
plt.show()
```
